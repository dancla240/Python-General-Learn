### Groupby och Groupby-object
- man kan göra **groupby** och sedan kolla på de olika grupperna som man skapat, med `groupbyobject.groups` . Det blir som en dict {grupp1 : [alla index i grupp 1], grupp2 : [alla index i grupp 2]}

```python
sessions = df.groupby('profile_id')
sessions.groups
```

Man kan få alla **keys** som finns i groupby-objeket:

```python
sessions.groups.keys()
```

Returnera alla index som hör till en viss grupp i groupby-objektet:

```python
sessions.groups
```

Man kan också inspektera de olika grupperna i det skapade groupby-objektet:

```python
sessions.get_group(2)
```

Beräkna medel ambient temperatur för varje grupp i groupby-objektet:

```python
mean_ambient = sessions['ambient'].mean()
display(mean_ambient)
```

Beräkna medel ambient och max ambient temperatur för varje grupp i groupby objektet:

```python
mean_max_ambient = sessions['ambient'].agg(['mean', 'max'])
display(mean_max_ambient)
```

### Aggregering
- använder -agg() funktionen
- ex: beräkna medelvärdet för 'pm' för profile_id == 18

```python
df[df['profile_id'] == 18]['pm'].agg('mean')
```

- beräkna även std och max

```python
df[df['profile_id'] == 18]['pm'].agg(['mean', 'std','max'])
```

Skapa en ny dataframe som innehåller allt där profile_id == 18.

```python
df_group_18 = df[df['profile_id'] == 18]
display(df_group_18)
```
