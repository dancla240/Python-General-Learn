## MongoDB
- MongoDB är en NoSQL dokument databas. I en **database** finns **collections**, och i en collection sparas **documents**.  
- **Collections** liknar tabeller i en SQL-databas.
- **Fields**: Tex "name" och "_id" i ett document, är ett **field**
- NoSQL lagrar datan som key:value pairs. Man kan också säga "non-tabular" database. Ett annat sätt att säga, är att de är "schema-less", dvs varje item behöver inte ha samma struktur. Detta är en stor fördel om datan hela tiden förändras.
- NoSQL kan skalas enklare än SQL databaser
- **hash** kopplar på något sätt key till value. Det gäller även för Python dictionaries.

### SQL

- I en relationsdatabas så definierar man tabellernas stuktur och varje item måste följa den strukturen.

### Allmänt
- **CRUD** operations are the four fundamental operations performed on a database: **Create**, **Read**, **Update**, and **Delete**.  
- **ACID**:
    - **Atomicity**: All or nothing
    - **Consistency**: Maintains database integrity
    - **Isolation**: Transactions do not interfere with each other
    - **Durability**: Changes are permanent


### MongoDB Queries
#### Basics
appdb> - betyder att databasen som heter appdb är aktiverad.  
`cls` - rensar skärmen  
`show dbs` - visar alla databaser  
`use databasnamn` - för att aktivera / skapa en ny databas. Man kan skapa en databas, men den skapas på riktigt först när man har adderat data till den.  
`show collections` - colletions = tabller inne i en databas  
`db.dropDatabase()` - ta bort databas  
Man accessar / aktiverar en databas genom att skriva namnet på databasen.
#### find
`db.users.find()` - finds everything within the database (eller inom "users" collection?)  
Allt man sparar är ett "document", som finns i en "collection" som finns i en "database".
#### insert
`db.users.insertOne({name: "John"})` - Addera till databasen, till "users" collection: insertOne är en funktion och  {name: "John"} är ett Java Script object, som man "pass to the function"  
`db.users.insertOne({name: "Sally", age: 19, address: {street:"987 North St}, hobbies: ["Running"]})` - ett annat sätt att addera mer komplicerad information till databasen  
`db.users.insertMany([{name: "Jill"}, {name: "Mike"}])` - adderar flera objekt till databasen samtidigt (typ en lista med de olika java script objekten)
#### find, sort and skip
`db.users.find().limit(2)` - hämtar information från databasen, men begränsar antalet svar till 2 st.  
`db.users.find().sort({name: 1})` - hämtar all information från databasen men svaret är sorterat på namn i alfabetiskt ordning  
`db.users.find().sort({name: -1})` - sorterat i omvänd alfabetisk ordning  
`db.users.find().sort({name: -1}).limit(2)` - samt begränsat antalet till 2  
`db.users.find().skip(1).limit(2)` - skip används för att hoppa över n stycken instances, så man börjar n steg ner i databasen
#### "where"
`db.users.find({name: "Kyle"})` - för att hitta allt där **field** "name" är "Kyle" (**where** name is Kyle.)
#### select
`db.users.find({name: "Kyle"}, {name: 1, age: 1, _id: 0})` - hittar alla med namnet Kyle, men returnear endast **fields** "name" och "age", samt returnerar inte "id" **field**.  
`db.users.find({name: "Kyle"}, {age: 0})` - hittar alla med namnet Kyle, returnerar all information föruton age  
#### mer komplicerad sökning
`db.users.find({name: {$eq: "Sally"}})` - istället för att söka direkt på "Sally" så pass ett nytt object { $eq: "Sally"}. Här kan man göra andra jämförelser såsom `$ne` not equal to, `$gt` greater than, `$gte` greater than or equal to, `$lte` less than or equal to, `$lt`less than.  
`db.users.find({name: {$in: ["Kyle", "Sally"]}})` - hitta allt i databasen vars namn finns i listan  
`db.users.find({age: {$exists: true}})` - hittar alla där det faktiskt finns ett field "age". Alla andra syns inte.  
`db.users.find({age: {$gte: 20, $lte: 40}, name: "Kyle"})` - hitta där **field** "age" är större än eller lika med 20 ***OCH*** mindre än eller lika med 40 ***OCH*** namnet är Kyle.  
#### AND / OR / NOT
`db.users.find({$and: [{age: 26}, {name: "Kyle"}]})` - AND villkor, fast detta går oftast att lösa på andra sätt.  
`db.users.find({or: [{age: {$lte: 20}}, {name: "Kyle"}]})` - OR villkor. Detta är enda sättet att göra OR.  
`db.users.find({age: {$not: {$lte: 20}}})` - NOT villkor
