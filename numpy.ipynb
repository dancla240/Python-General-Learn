```python
import numpy as np
```

```python
probabilities = np.random.rand(10, 10)
probabilities = probabilities / probabilities.sum(axis=0, keepdims=True)
probabilities
```

```python
accumulated_sum = np.cumsum(probabilities, axis=0)
accumulated_sum
```

```python
a = accumulated_sum[:, 0]
a
```

```python
index = np.where(a > 0.5)[0][0]
index
```
